# 1.7 Network graph visualization - 20th Century Data

```python
# Import Libraries
import pandas as pd
import numpy as np
import networkx as nx
import os
import matplotlib.pyplot as plt
from pyvis.network import Network
import seaborn as sns
import pyvis
```

```python
# Import relationship data
file_path = r"C:\Users\north\OneDrive\Dokumente\Career Foundry\Data Visualization 1\20th-Century\Data\country_relationships_summarized.csv"
relationship_df = pd.read_csv(file_path, index_col=0)
```

```python
relationship_df
```

```python
# Reset the index to make 'source' a regular column
relationship_df = relationship_df.reset_index()

# Reorder the columns to match the desired order: 'source', 'target', 'value'
relationship_df = relationship_df[['source', 'target', 'value']]

# Display the DataFrame to confirm changes
print(relationship_df.head())
```

```python
# Create a graph from the pandas dataframe
G = nx.from_pandas_edgelist(relationship_df, 
                            source="source", 
                            target="target", 
                            edge_attr="value", 
                            create_using=nx.Graph())
```

#### Graph visualization - Networkx

```python
# Plot the graph
plt.figure(figsize=(7, 7))
pos = nx.kamada_kawai_layout(G)
nx.draw(
    G,
    with_labels=True,
    node_color='red',
    edge_cmap=plt.cm.Blues,
    pos=pos
)
plt.title("Static Network Graph")

# Save the plot
plt.savefig("static_network_graph.png", dpi=300)
plt.show()
```

#### Graph visualization - Pyvis

```python
# Define net object

net = Network()
net = Network(notebook = True,width="900px", height="800px", bgcolor='#222222', font_color='white')

# Define nodes from the G object
node_degree = dict(G.degree)

# Setting up node size attributes
nx.set_node_attributes(G, node_degree, 'size')
net.from_nx(G)
net.show_buttons(filter_ = True)
net.repulsion()
net.show("20thcentury.html")
```

### Community detection - Leiden algorithm

```python
# Apply Leiden algorithm to detect communities
from cdlib import algorithms
coms = algorithms.leiden(G)
```

```python
# Convert the communities to a dictionary mapping nodes to their community
new_com = coms.to_node_community_map()

# Put the output into a dictionary where keys are nodes and values are their community
dict_com = {k: v[0] for k, v in new_com.items()}
```

```python
dict_com
```

```python
# Add community information to the NetworkX graph
nx.set_node_attributes(G, dict_com, 'group')

# Create a Pyvis Network graph
com_net = Network(notebook=False, width="1300px", height="700px", bgcolor="#222222", font_color="white")

# Load the NetworkX graph into Pyvis
com_net.from_nx(G)

# Add interactivity buttons and enable physics simulation
com_net.show_buttons(filter_="physics")
com_net.repulsion()

# Save the interactive graph as an HTML file
com_net.show("20thcentury_communities_leiden.html")

print("Interactive network graph with communities saved as '20thcentury_communities_leiden.html'.")
```

#### Observations on the Partition and Its Historical Relevance

1. **Does the Partition Make Sense According to History?**
   - The communities mostly align with historical events from the 20th century:
     - **Yellow Cluster (Central and Eastern Europe)**: Includes countries like Germany and Poland, reflecting their importance during World War I, World War II, and the Cold War.
     - **Red Cluster (Scandinavia)**: Denmark, Sweden, and Norway are grouped, showing their regional and cultural ties.
     - **Green Cluster (Asia and Pacific)**: Japan and Southeast Asian nations group together, reflecting colonial histories and World War II connections.
     - **Purple Cluster (United States and Others)**: The US is grouped with countries like Cuba, which may reflect Cold War or trade relationships.
       
2. **What Other Conclusions Can Be Made?**
   - **Central Countries**: Germany and France act as key hubs, connecting to multiple clusters, showing their historical importance.
   - **Regional Grouping**: Countries generally group based on geography or shared cultural and historical ties, such as Scandinavia and Southeast Asia.
   - **Cross-Community Links**: The US has connections to multiple groups, reflecting its global influence.

3. **Data Insights**:
   - Some groupings reflect historical relationships, while others (e.g., US with Cuba and Mozambique) might be influenced by trade, culture, or specific events in the dataset.

##### Conclusion
The communities make sense for the most part, reflecting historical events and regional ties. Key hubs like Germany and France show their importance, while cross-community links highlight global connections.

#### Centrality Measures

```python
# Calculate degree centrality
degree_dict = nx.degree_centrality(G)
degree_dict
```

```python
# Convert to a DataFrame for easy visualization
degree_df = pd.DataFrame(degree_dict.items(), columns=['countries', 'centrality'])

# Sort by centrality
degree_df.sort_values(by='centrality', ascending=False, inplace=True)
```

```python
# Plot degree centrality
plt.figure(figsize=(12, 16))  # Increase height for better visibility of all countries
sns.barplot(x="centrality", y="countries", data=degree_df, palette="Blues_d", saturation=0.9)
plt.title("20th Century - Degree Centrality", fontsize=16)
plt.xlabel("centrality", fontsize=14)
plt.ylabel("countries", fontsize=14)
plt.savefig("degree_centrality.png", dpi=300, bbox_inches="tight")
```

```python
# Calculate closeness centrality
closeness_dict = nx.closeness_centrality(G)

# Convert to a DataFrame for easy visualization
closeness_df = pd.DataFrame(closeness_dict.items(), columns=['countries', 'centrality'])

# Sort by centrality
closeness_df.sort_values(by='centrality', ascending=False, inplace=True)
```

```python
# Plot closeness centrality
plt.figure(figsize=(12, 16))
sns.barplot(x="centrality", y="countries", data=closeness_df, palette="Greens_d", saturation=0.9)
plt.title("20th Century - Closeness Centrality")
plt.savefig("closeness_centrality.png", dpi=300, bbox_inches="tight")
```

```python
# Calculate betweenness centrality
betweenness_dict = nx.betweenness_centrality(G)

# Convert to a DataFrame for easy visualization
betweenness_df = pd.DataFrame(betweenness_dict.items(), columns=['countries', 'centrality'])

# Sort by centrality
betweenness_df.sort_values(by='centrality', ascending=False, inplace=True)
```

```python
# Plot betweenness centrality
plt.figure(figsize=(12, 16))
sns.barplot(x="centrality", y="countries", data=betweenness_df, palette="Reds_d", saturation=0.9)
plt.title("20th Century - Betweenness Centrality")
plt.savefig("betweenness_centrality.png", dpi=300, bbox_inches="tight")
```

### Observations on Centrality Measures

#### **1. Degree Centrality**
- **Germany** has the highest degree centrality, indicating it has the most connections to other countries in the dataset.
- **Japan** and **France** follow, reflecting their strong connections, likely due to their economic or geopolitical importance in the 20th century.
- **Smaller countries** like **Lebanon**, **Mozambique**, and **Papua New Guinea** have very low degree centrality, indicating limited direct connections to others.

#### **2. Closeness Centrality**
- **Germany** again ranks highest, followed by **Japan** and **France**, suggesting these countries are well-positioned to access others efficiently in the network.
- Countries with lower closeness centrality, such as **Mozambique** and **Iraq**, are more isolated, requiring more steps to reach other countries in the network.

#### **3. Betweenness Centrality**
- **Germany** dominates in betweenness centrality, showing its role as a bridge between many other countries.
- **Japan**, **France**, and **India** also play significant bridging roles, likely reflecting their importance in connecting regions or clusters in the network.
- Countries with very low betweenness centrality, like **Lebanon** and **Latvia**, play minimal roles as intermediaries in the network.

---

```python
# Save centrality measures as node attributes in the graph
nx.set_node_attributes(G, degree_dict, 'degree_centrality')
nx.set_node_attributes(G, betweenness_dict, 'betweenness_centrality')
nx.set_node_attributes(G, closeness_dict, 'closeness_centrality')
```

```python
# Save the graph as a GraphML file
nx.write_graphml(G, "centrality_measures.graphml")
print("Graph saved as 'centrality_measures.graphml'.")
```

```python
# Create a DataFrame with centrality measures
centrality_data = []
for node, attrs in G.nodes(data=True):
    centrality_data.append({
        "Node": node,
        "Degree Centrality": attrs.get("degree_centrality", 0),
        "Betweenness Centrality": attrs.get("betweenness_centrality", 0),
        "Closeness Centrality": attrs.get("closeness_centrality", 0)
    })

df = pd.DataFrame(centrality_data)

# Save to CSV
df.to_csv("centrality_measures.csv", index=False)
print("Centrality measures saved to 'centrality_measures.csv'.")
```
